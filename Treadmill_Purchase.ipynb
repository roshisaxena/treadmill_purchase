```python
# Creating bins for Age - group

df.loc[df["Age"].between(18,19,"both"),"age-group"] = "Late teens"
df.loc[df["Age"].between(20,29,"both"),"age-group"] = "Twenties"
df.loc[df["Age"].between(30,39,"both"),"age-group"] = "Thirties"
df.loc[df["Age"].between(40,49,"both"),"age-group"] = "Forties"
df.loc[df["Age"].between(50,59,"both"),"age-group"] = "Fifties"

```

```python
# Creating bins for Income

df.loc[df["Income"].between(20000,29999,"both"),"income-group"] = "20K"
df.loc[df["Income"].between(30000,39999,"both"),"income-group"] = "30K"
df.loc[df["Income"].between(40000,49999,"both"),"income-group"] = "40K"
df.loc[df["Income"].between(50000,59999,"both"),"income-group"] = "50K"
df.loc[df["Income"].between(60000,69999,"both"),"income-group"] = "60K"
df.loc[df["Income"].between(60000,69999,"both"),"income-group"] = "60K"
df.loc[df["Income"].between(70000,79999,"both"),"income-group"] = "70K"
df.loc[df["Income"].between(80000,89999,"both"),"income-group"] = "80K"
df.loc[df["Income"].between(90000,99999,"both"),"income-group"] = "90K"
df.loc[df["Income"].between(100000,109999,"both"),"income-group"] = "100K"
```

```python
# Which Product is purchased by which age group

df1 = pd.crosstab(df["Product"],df["age-group"])
df1

```

```python
# What is the probability that the person in fifties will purchase a KP281 machine.

probability = (df1.loc["KP281"]["Fifties"])/(df1.loc["KP281"].sum())
print(probability)
```

```python
# What is the probability that the person in thirties will purchase a KP781 machine.

probability = (df1.loc["KP781"]["Thirties"])/(df1.loc["KP781"].sum())
print(probability)
```

```python
# What is the probability that the person in twenties will purchase a KP481 machine.

probability = (df1.loc["KP481"]["Twenties"])/(df1.loc["KP481"].sum())
print(probability)
```
